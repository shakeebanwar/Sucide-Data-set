# Visualize -  Number of Suicides - [ Selected Country ] - (Overall)

- Entered Years and Country,Number of suicides in each year will displayed by graph

```python
# --------------------------------------------------------------------------------------------------------------
                                                    #for Countries
# --------------------------------------------------------------------------------------------------------------
countriess=data.country.unique()
print("-------------------------------------------------------------------------------------------------------------------")
print("\t\t\t\t\tName of Attempted Suicides Countries")
print("-------------------------------------------------------------------------------------------------------------------")

for i in countriess:
    print(i,end=" --- ")

print("\n")

    
countries=input("Enter a Country name :  ")


print("\n")

# # --------------------------------------------------------------------------------------------------------------
#                                                         #for years
# # --------------------------------------------------------------------------------------------------------------
print("-----------------------------------------------------------------------------------------------------------------")
print(" \t\t\t\t\tYears of the Counties")
print("-----------------------------------------------------------------------------------------------------------------")
print("\t\t\t\t\tSelect the years of the Counntries to Start and End years")
print("-----------------------------------------------------------------------------------------------------------------")
info=data[(data.country==countries)]['year'].unique()
for i in info:
    print(i,end=' | ')

print("\n")

Start_year=int(input("\nEnter a Start Year , From Where You Want's :  "))
End_year=int(input("Enter an End year , From Where You End's : "))


print("\n")


# # --------------------------------------------------------------------------------------------------------------
#                                                         # Condition
# # --------------------------------------------------------------------------------------------------------------

detail=data[(data.country==countries) & ((data.year>=Start_year) & (data.year<=End_year)) & (data.suicides_no>=0)]
#print(detail)
year=detail.year
no_suicide=detail.suicides_no

# # ----------------------------------------------------------------------------------------------------------------
#                                                         #Graph
# # ----------------------------------------------------------------------------------------------------------------

plt.figure(figsize=(15,5))
plt.title(f"No Of Suicides {countries} in Selected Years",fontsize=20)
plt.xlabel("Suicides Years",fontsize=15)
plt.ylabel(f"Number of Suicides in {countries}",fontsize=15)
plt.bar(year,no_suicide,color='cornflowerblue')
plt.show()
#plt.legend(fontsize=15)

```

# Visualize - Number of Suicides - [ Selected Country ] - (Male / Female)
- Entered Years and Country,Number of suicides in gender Male or Female in each year will displayed by graph

```python
# --------------------------------------------------------------------------------------------------------------
                                                    #for Countries
# --------------------------------------------------------------------------------------------------------------
countriess=data.country.unique()
print("-------------------------------------------------------------------------------------------------------------------")
print("\t\t\t\t\tName of Attempted Suicides Coutries")
print("-------------------------------------------------------------------------------------------------------------------")

for i in countriess:
    print(i,end=" --- ")

print("\n")

    
countries=input("Enter a Country name :  ")


print("\n")

# # --------------------------------------------------------------------------------------------------------------
#                                                         #for years
# # --------------------------------------------------------------------------------------------------------------
print("-----------------------------------------------------------------------------------------------------------------")
print(" \t\t\t\t\tYears of the Counties")
print("-----------------------------------------------------------------------------------------------------------------")
print("\t\t\t\t\tSelect the years of the Counntries to Start and End years")
print("-----------------------------------------------------------------------------------------------------------------")
info=data[(data.country==countries)]['year'].unique()
for i in info:
    print(i,end=' | ')

print("\n")

Start_year=int(input("\nEnter a Start Year , From Where You Want's :  "))
End_year=int(input("Enter an End year , From Where You End's : "))
gender=input("Enter a gender : ")

print("\n")


# # --------------------------------------------------------------------------------------------------------------
#                                                         # Condition
# # --------------------------------------------------------------------------------------------------------------

detail=data[(data.country==countries) & ((data.year>=Start_year) & (data.year<=End_year)) & (data.suicides_no>=0) & (data.sex==gender)]
year=detail.year
no_suicide=detail.suicides_no

if gender=='male':
    plt.figure(figsize=(15,5))
    plt.title("Number of Suicide A/c to Gender - Male and Female\n",fontsize=20)
    plt.xlabel(f"Male-Sucides Attempt {countries} Selected Year",fontsize=15)
    plt.ylabel("Number of Suicides",fontsize=15)
    plt.bar(year,no_suicide,label="Male",width=0.8)
    plt.legend(fontsize=15)
else:
    plt.figure(figsize=(15,5))
    plt.title(f"Number of Suicide A/c to Gender - Male and Female in {countries}",fontsize=20)
    plt.xlabel(f"Female-Sucides Attempt in {countries} Selected Year",fontsize=15)
    plt.ylabel("Number of Suicides",fontsize=15)
    plt.bar(year,no_suicide,label='Female',width=0.8)
    
    plt.legend(fontsize=15) #legend jb use karaga jb labels hga 

    


```
